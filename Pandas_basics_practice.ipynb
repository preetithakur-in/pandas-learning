```python
'''*Q7.* Given a DataFrame of daily closing prices for AAPL over 10 days (make up the values), calculate:
- *Daily returns* using .pct_change()
- *Cumulative returns* using .cumprod()
- *20-day rolling mean* using .rolling()'''
```

```python
aalp = pd.DataFrame({'close':[234,653,836,836,832,368,845,463,836,737]})

aalp['daily_return'] = aalp['close'].pct_change()

aalp['cum_return'] = (1+ aalp['daily_return'].fillna(0)).cumprod()

aalp['rolling_mean']= aalp['close'].rolling(window=3).mean()

aalp
```

```python
'''*Q8.* You have two DataFrames:
python
prices_df  # Contains: Date, Ticker, Close
returns_df # Contains: Date, Ticker, Return

*Merge* them on both Date and Ticker columns into a single DataFrame.'''
```

```python
prices_df = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03'],
    'Ticker': ['AAPL', 'GOOGL', 'AAPL', 'GOOGL', 'AAPL'],
    'Close': [185.2, 140.5, 187.8, 142.1, 186.5]
})

returns_df = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03'],
    'Ticker': ['AAPL', 'GOOGL', 'AAPL', 'GOOGL', 'AAPL'],
    'Return': [0.012, -0.005, 0.014, 0.011, -0.007]
})

merged_df = prices_df.merge(returns_df, on=['Date', 'Ticker'])
merged_df
```

```python
'''*Q9.* Given a DataFrame with daily returns for 3 stocks (AAPL, TSLA, MSFT), *group by Ticker* and calculate:
- Mean return
- Max return
- Volatility (std)

All in *one operation* using .groupby() and .agg().'''
```

```python
df = pd.DataFrame({
    'Ticker': ['AAPL', 'AAPL', 'AAPL', 'TSLA', 'TSLA', 'TSLA', 'MSFT', 'MSFT', 'MSFT'],
    'Return': [0.012, -0.005, 0.008, 0.025, -0.018, 0.031, 0.008, 0.003, -0.004]
})

print("Original DataFrame:\n", df)

# Group by Ticker and aggregate
result = df.groupby('Ticker')['Return'].agg(
    Mean_Return='mean',
    Max_Return='max',
    Volatility='std'
).round(4)

print("\nResult:\n", result)
```

```python
'''*Q10.* Given a DataFrame of monthly stock prices for a full year (12 rows),
*resample it to quarterly* frequency and calculate the
*average price per quarter* using .resample().'''
```

```python
df = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=12, freq='MS'),
    'Price': [150, 155, 160, 158, 162, 170, 175, 172, 168, 165, 171, 180]
})

df.set_index('Date', inplace=True)

print("Monthly Data:\n", df)

result = df['Price'].resample('QE').mean()

print("\nQuarterly Average Price:\n", result)
```
